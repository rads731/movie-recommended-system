```python
import numpy as np
import pandas as pd
import warnings
import json
from sklearn.feature_extraction.text import TfidfVectorizer,CountVectorizer
from sklearn.metrics.pairwise import linear_kernel, cosine_similarity

movie_df=pd.read_csv(r'C:\Users\nEW u\Desktop\data science\data sets\movie_data\tmdb_5000_movies.csv')
credit_df=pd.read_csv(r'C:\Users\nEW u\Desktop\data science\data sets\movie_data\tmdb_5000_credits.csv')
df=pd.merge(movie_df,credit_df)
genres={}
s=[]
for i in range(0,4809):
    genres=json.loads(df.genres.loc[i])
    s.append(genres)
    df2=pd.DataFrame(s)

gen=[]
d=[]
for i in range(0,4809):
    d=[]
    for j in range(0,7):
        if df2.iloc[i,j]!=None:
            di=df2.iloc[i,j]
            d.append(di['name'])
    gen.append(d)     
    

df3=pd.DataFrame(gen)
df3['movie_id']=df['movie_id']
gen=[]

d=""
df.insert(2,'combined_genre',"")
for i in range(0,4809):
    d=""
    for j in range(0,7):
        if df3.loc[i,j]==None:
            continue
        d=d+df3.loc[i,j]+' '
    df.combined_genre.loc[i]=d

df.insert(3,'bag_of_words',"")
df_test=pd.DataFrame(df['combined_genre'])
columns = df_test.combined_genre
d=""
for i in range(0,4809):
    d=""
        #if df_test.iloc[i,j]==None:
        #    continue
    d=d+df_test.iloc[i,0]+' '
    df.bag_of_words.loc[i]=d

df_test=df[['title','bag_of_words']]
df_test.set_index('title',inplace=True)

count = CountVectorizer()
count_matrix = count.fit_transform(df_test['bag_of_words'])

indices = pd.Series(df_test.index)
indices[:5]

cosine_sim = cosine_similarity(count_matrix, count_matrix)
def recommendations(title, cosine_sim = cosine_sim):
    
    recommended_movies = []
   # ids=df[df.title==title].movie_id
    # gettin the index of the movie that matches the title
    idx = indices[indices == title].index[0]

    # creating a Series with the similarity scores in descending order
    score_series = pd.Series(cosine_sim[idx]).sort_values(ascending = False)

    # getting the indexes of the 10 most similar movies
    indexes = list(score_series.iloc[1:100].index)
    
    # populating the list with the titles of the best 10 matching movies
    for i in indexes:
        recommended_movies.append(list(df_test.index)[i])
        
    return recommended_movies

recommendations('Batman')
```
